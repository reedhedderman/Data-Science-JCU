```python
import math as math
import matplotlib.pyplot as plt
import numpy as np

class QuadraticFunction:

    def __init__(self) -> None:
        pass

    def client_input(self):
        self.a,self.b,self.c = input("Write the values of your coefficient: ").split()
        self.a = float(self.a)
        self.b = float(self.b)
        self.c = float(self.c)

        print("Coefficients a, b, and c are: {0:.2f}, {1:.2f}, {2:.2f}".format (self.a,self.b,self.c))
       

    def x_intercepts(self):
        discrimanant = math.pow(self.b,2) - 4*self.a*self.c

        if(discrimanant >=0):
            root_1 = ((-self.b+(math.sqrt(discrimanant)))/(2*self.a))
            root_2 = ((-self.b-(math.sqrt(discrimanant)))/(2*self.a))
            #output = True
            print("x-intercepts: ({0:.3f},0), ({1:.3f},0)".format(root_1,root_2))
        else:
            print("quadratic function does not have an x-intercept")
            #ouput = False

        #return output

    def vertex(self):
        h = -self.b / (2 * self.a)
        k = (self.a * (math.pow(h,2)) + self.b * h + self.c)
        print("vertex: ({0:.3f},{1:.3f})".format(h,k))

    def plot_QuadraticFunction(self):
        #plot your function

        #define you x space
        x = np.linspace(-40, 40, 100)

        #define your quadratic finction

        y = self.a * (x**2) + self.b * x + self.c

        fig = plt.figure()
        ax = fig.add_subplot(1,1,1)
        ax.spines['left'].set_position("center")
        ax.spines['bottom'].set_position("center")
        ax.spines['right'].set_color("none")
        ax.spines['top'].set_color("none")
        ax.xaxis.set_ticks_position("bottom")
        ax.yaxis.set_ticks_position("left")

        plt.plot(x, y, 'r')

```

```python
Poly = QuadraticFunction()
Poly.client_input()
Poly.vertex()
Poly.x_intercepts()
Poly.plot_QuadraticFunction()
```
