```python
import math as math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def client_input():
    a,b,c = input("Write the values of your coefficient: ").split()
    a = float(a)
    b = float(b)
    c = float(c)

    print("Coefficients a, b, and c are: {0:.2f}, {1:.2f}, {2:.2f}".format(a,b,c))
    return a,b,c 

def x_intercepts(a,b,c):
    discrimanant = math.pow(b,2) - 4*a*c

    if(discrimanant >=0):
        root_1 = ((-b+(math.sqrt(discrimanant)))/(2*a))
        root_2 = ((-b-(math.sqrt(discrimanant)))/(2*a))
        #output = True
        print("x-intercepts: ({0:.3f},0), ({1:.3f},0)".format(root_1,root_2))
    else:
        print("quadratic function does not have an x-intercept")
        #ouput = False

    #return output

def vertex(a,b,c):
    h = -b / (2 * a)
    k = (a * (math.pow(h,2)) + b * h + c)
    print("vertex: ({0:.3f},{1:.3f})".format(h,k))

def plot_QuadraticFunction(a,b,c):
    #plot your function

    #define you x space
    x = np.linspace(-40, 40, 100)

    #define your quadratic finction

    y = a * (x**2) + b * x + c

    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.spines['left'].set_position("center")
    ax.spines['bottom'].set_position("center")
    ax.spines['right'].set_color("none")
    ax.spines['top'].set_color("none")
    ax.xaxis.set_ticks_position("bottom")
    ax.yaxis.set_ticks_position("left")

    plt.plot(x, y, 'r')

```

```python
#main function


a,b,c = client_input()

vertex(a,b,c)
x_intercepts(a,b,c)
plot_QuadraticFunction(a,b,c)

    
```
